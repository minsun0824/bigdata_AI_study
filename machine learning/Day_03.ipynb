# ML 코딩
       1. 모델을 생성: model=머신러닝 알고리즘(hyper parameter) <-클래스

       4. 예측: model.predict(X_new)

```python
train_accuracy=[]
test_accuracy=[]
for n_neighbors in range(1,11): #1부터 10까지 네이버 늘림
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    #장점은 이해하기 쉽다. 단점은 튜닝하기 어렵다
    clf.fit(X_train,y_train)
    train_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
```

### SVC의 규제:C,cost
- C가 클수록 돈이 많아서 규제가 적다고 생각 가능
- 규제가 작으면 오버피팅을 줄이지만 언더피팅 발생 가능, 규제가 크면 그 반대

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(
    cancer.data, cancer.target,stratify=cancer.target,random_state=42)
```

```python
logreg=LogisticRegression().fit(X_train,y_train)
print('train accuracy:{:.3f}'.format(logreg.score(X_train,y_train)))
print('test accuracy:{:.3f}'.format(logreg.score(X_test,y_test)))
```

```python
logreg001=LogisticRegression(C=0.01).fit(X_train,y_train)
print('train accuracy:{:.3f}'.format(logreg001.score(X_train,y_train)))
print('test accuracy:{:.3f}'.format(logreg001.score(X_test,y_test)))
```

```python
#100으로 하면 안됨. 줄여보자
```

```python
logreg100=LogisticRegression(C=100).fit(X_train,y_train)
print('train accuracy:{:.3f}'.format(logreg100.score(X_train,y_train)))
print('test accuracy:{:.3f}'.format(logreg100.score(X_test,y_test)))
```

```python
#0.01로 규제하는게 정답
for C,marker in zip([0.001,1,100],['o','^','v']): #그려주기 옵션
    lr_l1=LogisticRegression(solver='liblinear',C=C, penalty='l1').fit(X_train,y_train)
    print('C:{:.3f}, 훈련 정확도:{:.2f}'.format(C,lr_l1.score(X_train,y_train)))
    print('C:{:.3f}, 테스트 정확도:{:.2f}'.format(C,lr_l1.score(X_test,y_test)))
    
#3개 비교 시 100이 정확하다.똑같이 안나오긴 함 ㅠ 
```

```python
# ensemble: 성능이 낮은 트리 여러 개를 합쳐서 성능을 올리는 기법. 
# 집단지성 같은 느낌.
from sklearn.tree import DecisionTreeClassifier
cancer=load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(
    cancer.data, cancer.target,stratify=cancer.target,random_state=42)
tree=DecisionTreeClassifier(random_state=0)
#훈련.
#학습: 틀린거를 줄여가는 과정
tree.fit(X_train,y_train)
print('훈련정확도:{:3f}'.format(tree.score(X_train,y_train)))
print('테스트정확도:{:3f}'.format(tree.score(X_test,y_test)))

#훈련정확도:1.000000 ->백퍼센트: overfitting 의사결정 트리는 과정을 거치면서 depth를 외워버리므로 오버
#피팅을 줄일려먼 depth 길이를 제한해주어야 함
#3까지 제한하면 3이후의 과정을 외우지 않음ㅠㅠ
#테스트정확도:0.937063
```

### Maxdepth: 의사결정 트리에서 overfitting줄이는 방법
훈련정확도:1.000000 ->백퍼센트: overfitting 의사결정 트리는 과정을 거치면서 depth를 외워버리므로 오버피팅을 줄일려먼 depth 길이를 제한해주어야 함

3까지 제한하면 3이후의 과정을 외우지 않음



```python
tree=DecisionTreeClassifier(max_depth=4,random_state=0)
tree.fit(X_train,y_train)
print('훈련정확도:{:3f}'.format(tree.score(X_train,y_train)))
print('테스트정확도:{:3f}'.format(tree.score(X_test,y_test)))
```

### 중요도: 의사결정 트리에서 중요하게 생각하는 것

```python
print('특성중요도',tree.feature_importances_) #밑에 밑줄은 read only 데이터이므로 고치치 말아라
```

```python
plt.figure(figsize=(15,10))
def plot_feature_importances_cancer(model):
    n_features=cancer.data.shape[1] #0번째는환자수, 1은 특징 수
    plt.barh(np.arange(n_features),model.feature_importances_,align='center')
    plt.yticks(np.arange(n_features),cancer.feature_names)
    plt.xlabel('feature importance') #x축
    plt.ylabel('feature') #y축
    plt.ylim(-1,n_features)

plot_feature_importances_cancer(tree)
```
